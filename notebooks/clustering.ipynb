```python
%matplotlib inline

import pandas as pd
```

```python
head = 1600
matrix_file = '../data/GSE60755_CE_exp.txt'


matrix = pd.read_csv(matrix_file,delimiter='\t')
matrix = matrix.head(head)
```

```python
import sklearn.manifold as sklm
mvalues = matrix.values[:,1:]
model = sklm.TSNE(metric='cosine')

f = model.fit_transform(mvalues)
```

```python
import matplotlib.pyplot as plt

plt.plot(f[:,0], f[:,1], 'o')
```
